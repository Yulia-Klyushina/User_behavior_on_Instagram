**Analysing User Behaviour on Instagram using Python and SQL.**

This project includes analyzing a dataset of user behavior of a highly popular social media platform, Instagram. This project aims on cleaning the dataset, analyze the given dataset, and mine informational quality insights using Python and SQL.

**Data preperation**

Removing Unwanted Columns and Renaming Appropriate Columns for Improved Analysis. Data Cleaning is a process of removing unwanted columns and renaming the appropriate columns in a dataset. It involves identifying and removing irrelevant or redundant columns and renaming the columns to make them more descriptive and consistent with the content of the dataset. This improves the quality and usability of the data for analysis and modeling purposes.

```python
import pandas as pd
import numpy as np
```

```python
comments=pd.read_csv('comments.csv')
comments.head()
```

```python
#Remove Unwanted columns in Comments table
comments = comments.drop(['posted date', 'emoji used', 'Hashtags used count'], axis=1)    
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	id
    # 2.	comment_text
    # 3.	user_id
    # 4.	photo_id
    # 5.	created_at    
    
comments = comments.rename(columns={
                        'comment': 'comment_text',
                        'User  id': 'user_id',
                        'Photo id': 'photo_id',
                        'created Timestamp': 'created_at'})
comments.head()
```

```python
#export cleaned Dataset to new csv file named "comments_cleaned.csv"
comments.to_csv('comments_cleaned.csv')
```

```python
#read follows dataset
follows =pd.read_csv('follows.csv')
follows.head()
```

```python
#Remove Unwanted columns
follows = follows.drop(['is follower active', 'followee Acc status'], axis=1)
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	follower_id
    # 2.	followee_id
    # 3.	created_at
follows = follows.rename(columns={
                        'follower': 'follower_id',
                        'followee ': 'followee_id',
                        'created time': 'created_at'})
```

```python
#export cleaned Dataset to new csv file named "follows_cleaned.csv"
follows.to_csv('follows_cleaned.csv')   
```

```python
#read likes dataset
likes=pd.read_csv('likes.csv')
likes.head()
```

```python
#Remove Unwanted columns
likes = likes.drop(['following or not', 'like type'], axis=1)
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	user_id
    # 2.	photo_id
    # 3.	created_at
likes = likes.rename(columns={
                        'user ': 'user_id',
                        'photo': 'photo_id',
                        'created time': 'created_at'})    
```

```python
#export cleaned Dataset to new csv file named "likes_cleaned.csv"
likes.to_csv('likes_cleaned.csv')
```

```python
#read photo_tags dataset
photo_tags=pd.read_csv('photo_tags.csv')
photo_tags.head()
```

```python
#Remove Unwanted columns
photo_tags = photo_tags.drop(['user id'], axis=1)
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	photo_id
    # 2.	tag_id
photo_tags = photo_tags.rename(columns={
                        'photo': 'photo_id',
                        'tag ID': 'tag_id'})
```

```python
 #export cleaned Dataset to new csv file named "photo_tags_cleaned.csv"
photo_tags.to_csv('photo_tags_cleaned.csv')
```

```python
#read photos dataset
photos=pd.read_csv('photos.csv')
photos.head()
```

```python
#Remove Unwanted columns
photos = photos.drop(['Insta filter used', 'photo type'], axis=1)
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	id
    # 2.	image_url
    # 3.	user_id
    # 4.	created_date
photos = photos.rename(columns={
                        'image link': 'image_url',
                        'user ID': 'user_id',
                        'created dat': 'created_date'})
```

```python
#export cleaned Dataset to newcsv file named "photos_cleaned.csv"
photos.to_csv('photos_cleaned.csv')
```

```python
#read tags dataset
tags=pd.read_csv('tags.csv')
tags.head()
```

```python
#Remove Unwanted columns
tags = tags.drop(['location'], axis=1)
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	id
    # 2.	tag_name
    # 3.	created_at
tags = tags.rename(columns={
                        'tag text': 'tag_name',
                        'created time': 'created_at'})
```

```python
#export cleaned Dataset to newcsv file named "tags_cleaned.csv"
tags.to_csv('tags_cleaned.csv')
```

```python
#read users dataset
users=pd.read_csv('users.csv')
users.head()
```

```python
#Remove Unwanted columns
users = users.drop(['private/public', 'post count', 'Verified status'], axis=1)
```

```python
#rename columns, only these columns are allowed in the dataset
    # 1.	id
    # 2.	username
    # 3.	created_at
users = users.rename(columns={
                        'name': 'username',
                        'created time': 'created_at'})
```

```python
#export cleaned Dataset to newcsv file named "users_cleaned.csv"
users.to_csv('users_cleaned.csv')
```
